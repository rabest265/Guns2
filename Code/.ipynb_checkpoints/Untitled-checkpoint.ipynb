```python

# Dependencies
import requests
import json
from config import gkey
import os
import csv
import pandas as pd
import numpy as np
```

```python
# Load CSV file
csv_path = os.path.join('..',"output", "cities.csv")

# Read Purchasing File and store into Pandas data frame
cities_df = pd.read_csv(csv_path, encoding = "ISO-8859-1")
cities_df.head()
```

```python

cities_df["Lat"] = ""
cities_df["Lng"] = ""

cities_df.head()
```

```python

cities_df['county'] = cities_df.City.str.extract('.*\((.*)\).*')
cities_df['county'] = cities_df.county.str.rsplit(" ",1).str[0]

cities_df['cityarea'] = cities_df.City.str.rsplit(" (",1).str[0]

cities_df['city_name'] = cities_df.cityarea.str.rsplit(" ",1).str[0]

cities_df['city_type'] = cities_df.cityarea.str.split().str[-1]

cities_df.head(10)
```

```python
params = {"key": gkey}
cnt=0
# Loop through the cities_df and run a lat/long search for each city

for index, row in cities_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    city = row['city_name']
    state = row['State']
    citystate=city + ", "+ state
    cnt=cnt+1
    if cnt<5000:
        print (citystate)
        
    # update address key value
    params['address'] = f"{city},{state}"

    # make request
    cities_lat_lng = requests.get(base_url, params=params)

    # print the cities_lat_lng url, avoid doing for public github repos in order to avoid exposing key
#    print(cities_lat_lng.url)

    # convert to json
    cities_lat_lng = cities_lat_lng.json()

    cities_df.loc[index, "Lat"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
    cities_df.loc[index, "Lng"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]

# Visualize to confirm lat lng appear
cities_df.head()
```

```python
cities_df.head()
cities_df.to_csv("../Output/cities.csv", index=False, header=True)
```
