```python

# Dependencies
import requests
import json
from config import gkey
import os
import csv
import pandas as pd
import numpy as np
```

```python
# Load CSV file
csv_path = os.path.join('..',"data", "demographics.csv")

# Read Purchasing File and store into Pandas data frame
cities_df = pd.read_csv(csv_path, encoding = "ISO-8859-1")
cities_df.head()
```

```python

cities_df["Lat"] = ""
cities_df["Lng"] = ""

cities_df.head()
```

```python
#Extract County, City Name and City Type

cities_df['county'] = cities_df.City.str.extract('.*\((.*)\).*')
cities_df['county'] = cities_df.county.str.rsplit(" ",1).str[0]

cities_df['cityarea'] = cities_df.City.str.rsplit(" (",1).str[0]

cities_df['city_name'] = cities_df.cityarea.str.rsplit(" ",1).str[0]

cities_df['city_type'] = cities_df.cityarea.str.split().str[-1]

cities_df.head(10)
```

```python
#Save to CSV file

cities_df.head()
cities_df.to_csv("../Output/demographics.csv", index=False, header=True)
```
