```python
csv_path = "data/gun-violence-data_01-2013_12-2015.csv"
gunviolence_df = pd.read_csv(csv_path)
gunviolence_df.head()
```

```python
text = gunviolence_df.incident_characteristics[0]
```

```python
wordcloud = WordCloud().generate(text)
```

```python
wordcloud = WordCloud(width=1600, height=800).generate(text)
# Open a plot of the generated image.

plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.savefig('images/final_wordcloud')
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=1000, background_color="black").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('images/output.png')
plt.show()

wordcloud = WordCloud(max_font_size= 200, max_words=1000, background_color="white", width=1600, height=800).generate(text)


plt.figure( figsize=(20,10), facecolor='k')
#plt.imshow(wordcloud)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.savefig('images/final_wordcloud')
plt.show()

```

```python
gun_mask = np.array(Image.open("images/gun_mask1.png"))

gun_mask = gun_mask.reshape((gun_mask.shape[0],-1), order='F')

```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
transformed_gun_mask = np.ndarray((gun_mask.shape[0],gun_mask.shape[1]), np.int32)

for i in range(len(gun_mask)):
    transformed_gun_mask[i] = list(map(transform_format, gun_mask[i]))
    
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=transformed_gun_mask,
                contour_width=3, contour_color='firebrick')
```

```python
wc.generate(text)
```

```python
wc.to_file("gun_word_cloud.png")
```

```python
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
